```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('/kaggle/input/automobile-dataset/Automobile.csv')
```

```python
# Check for missing values and drop rows with NaN values
df = df.dropna()
```

```python
# Separate the features and target variable
X = df.drop(['mpg'], axis=1)
y = df['mpg']
```

```python
# One-hot encode categorical columns
X_encoded = pd.get_dummies(X, columns=['name', 'origin'])
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Perform standardization on numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define the model architecture
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test_scaled)
```

```python
# Flatten the arrays
y_test = np.ravel(y_test)
y_pred = np.ravel(y_pred)
```

```python
# Compute mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error: {:.2f}".format(mse))
```

```python
# Compute mean absolute error (MAE)
mae = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error: {:.2f}".format(mae))
```

```python
# Compute R-squared (coefficient of determination)
r2 = r2_score(y_test, y_pred)
print("R-squared: {:.2f}".format(r2))
```
